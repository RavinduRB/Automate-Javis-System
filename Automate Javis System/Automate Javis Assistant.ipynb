```python
import threading
import speech_recognition as sr
import spacy
import pyttsx3
import smtplib
from email.mime.text import MIMEText
import requests
import psutil
import time
import cv2
from deepface import DeepFace
import tkinter as tk

def recognize_speech_from_mic(recognizer, microphone):
    while True:
        with microphone as source:
            print("Listening for commands...")
            audio = recognizer.listen(source)
        
        try:
            response = recognizer.recognize_google(audio)
            print(f"Command: {response.lower()}")
            if "hey jarvis" in response.lower():
                print("Wake word detected! Listening for further commands...")
                # Further processing
        except sr.RequestError:
            print("API unavailable")
        except sr.UnknownValueError:
            print("Unable to recognize speech")

# Initialize recognizer and microphone
recognizer = sr.Recognizer()
mic = sr.Microphone()

# Run voice recognition in a separate thread
voice_thread = threading.Thread(target=recognize_speech_from_mic, args=(recognizer, mic))
voice_thread.start()

nlp = spacy.load("en_core_web_sm")

def process_command(command_text):
    doc = nlp(command_text)
    intents = []
    entities = []
    
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    if "reminder" in command_text:
        intents.append("set_reminder")
    elif "email" in command_text:
        intents.append("send_email")
    
    print("Intents:", intents)
    print("Entities:", entities)

def handle_command(command_text):
    nlu_thread = threading.Thread(target=process_command, args=(command_text,))
    nlu_thread.start()

# Example usage
command_text = "Set a meeting with John tomorrow at 3 PM."
handle_command(command_text)

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def threaded_speech(text):
    speech_thread = threading.Thread(target=speak, args=(text,))
    speech_thread.start()

# Example usage
threaded_speech("Hello, I am JARVIS. How can I assist you today?")

def send_email(to_email, subject, body):
    from_email = "your_email@gmail.com"
    password = "your_password"
    
    msg = MIMEText(body)
    msg["Subject"] = subject
    msg["From"] = from_email
    msg["To"] = to_email

    server = smtplib.SMTP_SSL("smtp.gmail.com", 465)
    server.login(from_email, password)
    server.sendmail(from_email, to_email, msg.as_string())
    server.quit()

def threaded_email(to_email, subject, body):
    email_thread = threading.Thread(target=send_email, args=(to_email, subject, body))
    email_thread.start()

# Example usage
threaded_email("recipient@example.com", "Meeting Reminder", "Don't forget about the meeting at 3 PM.")

def control_smart_device(entity_id, action):
    url = "http://your-home-assistant-url/api/services/homeassistant/turn_on"
    headers = {
        "Authorization": "Bearer your_access_token",
        "Content-Type": "application/json"
    }
    data = {
        "entity_id": entity_id
    }
    if action == "off":
        url = "http://your-home-assistant-url/api/services/homeassistant/turn_off"
    
    response = requests.post(url, headers=headers, json=data)
    if response.status_code == 200:
        print("Device control successful")
    else:
        print("Failed to control device")

def threaded_smart_home(entity_id, action):
    smart_home_thread = threading.Thread(target=control_smart_device, args=(entity_id, action))
    smart_home_thread.start()

# Example usage
threaded_smart_home("light.living_room", "on")

def monitor_system():
    while True:
        cpu_usage = psutil.cpu_percent()
        memory_info = psutil.virtual_memory()
        disk_info = psutil.disk_usage('/')
        
        print(f"CPU Usage: {cpu_usage}%")
        print(f"Memory Usage: {memory_info.percent}%")
        print(f"Disk Usage: {disk_info.percent}%")
        
        time.sleep(5)  # Pause for 5 seconds before checking again

# Run system monitoring in a separate thread
monitoring_thread = threading.Thread(target=monitor_system)
monitoring_thread.start()

def recognize_face():
    video_capture = cv2.VideoCapture(0)
    
    while True:
        ret, frame = video_capture.read()
        cv2.imshow('Video', frame)
        
        try:
            face_data = DeepFace.analyze(frame, actions=['emotion'])
            print(face_data)
        except Exception as e:
            print("Face recognition error:", e)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    video_capture.release()
    cv2.destroyAllWindows()

# Run facial recognition in a separate thread
face_recognition_thread = threading.Thread(target=recognize_face)
face_recognition_thread.start()

def send_command():
    user_input = entry.get()
    print(f"Command received: {user_input}")
    # Further processing in a separate thread
    threading.Thread(target=process_command, args=(user_input,)).start()

root = tk.Tk()
root.title("JARVIS Assistant")

entry = tk.Entry(root, width=50)
entry.pack()

button = tk.Button(root, text="Send", command=send_command)
button.pack()

# Run the GUI mainloop
root.mainloop()

```
